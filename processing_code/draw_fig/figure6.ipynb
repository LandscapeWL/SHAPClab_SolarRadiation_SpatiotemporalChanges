```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import csv

# 设置默认字体
plt.rcParams['font.sans-serif']=['Times New Roman'] 
```

```python
# 读取数据

filename = '6.2019-2013_radiation.csv'  # 更改为你的CSV文件名
column_a_data = []

with open(filename, newline='', encoding='utf-8') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    for row in csv_reader:
        column_a_data.append(row['2019-2013'])

# 将字符串数据转换为浮点数
column_a_data = [float(item) for item in column_a_data]

print(column_a_data)
```

```python
# 2013
# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(5,4),dpi=300)

# 设置y轴的刻度范围，使正负数值都能显示
ax.set_ylim(-3000, 2000)

# 使用 ggplot 样式
plt.style.use('ggplot')
# 将背景设为白色
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# 设置边界线颜色为黑色
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['top'].set_color('black')
ax.spines['right'].set_color('black')

# 绘制正值柱状图
positive_pointsx = [x for x in column_a_data if x >= 0]
n, bins, patches = ax.hist(positive_pointsx, bins=60, density=False, alpha=0.75, rwidth=0.8, color='#ea3323')

# 绘制负值柱状图
negative_pointsx = [x for x in column_a_data if x < 0]
n, bins, patches = ax.hist(negative_pointsx, bins=60, density=False, alpha=0.75, rwidth=0.8, color='#58a530')

# 设置负值的y轴坐标为负
for patch, value in zip(patches, negative_pointsx):
    patch.set_y(-patch.get_height())

# 设置轴标签
plt.xlabel('Street view quantity statistics', color='black', fontsize=10)
plt.ylabel('Variation in solar radiation(wh/m2)', color='black', fontsize=10)

# 设置标题和坐标轴标签
# ax.set_title('Perception Score Distribution of Difference')

# 设置y轴刻度范围为0-3000
# ax.set_ylim(0, 3000)

# 自定义坐标轴刻度字体大小和颜色
ax.tick_params(axis='both', labelsize=10, colors='black')

# 显示图形
plt.show()
```
