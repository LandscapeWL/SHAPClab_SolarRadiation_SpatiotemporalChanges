```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

# 设置默认字体
plt.rcParams['font.sans-serif']=['Times New Roman'] 
```

```python
# 读取一个shp 或者geojson文件
points = gpd.read_file(r'/Users/wl/Phd_Paper/202301Perception_Historical/gis_data/point_2019-2013.shp')
points
```

```python
pointsx = list(points['difference'])
```

```python
# 2013
# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(6,4),dpi=300)

# 设置y轴的刻度范围，使正负数值都能显示
ax.set_ylim(-2500, 2500)

# 使用 ggplot 样式
plt.style.use('ggplot')
# 将背景设为白色
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# 设置边界线颜色为黑色
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['top'].set_color('black')
ax.spines['right'].set_color('black')

# 绘制正值柱状图
positive_pointsx = [x for x in pointsx if x >= 0]
n, bins, patches = ax.hist(positive_pointsx, bins=60, density=False, alpha=0.75, rwidth=0.8, color='#e66f51')

# 绘制负值柱状图
negative_pointsx = [x for x in pointsx if x < 0]
n, bins, patches = ax.hist(negative_pointsx, bins=60, density=False, alpha=0.75, rwidth=0.8, color='#287271')
# 设置负值的y轴坐标为负
for patch, value in zip(patches, negative_pointsx):
    patch.set_y(-patch.get_height())

# 设置标题和坐标轴标签
ax.set_title('Perception Score Distribution of Difference', fontsize=16)
# ax.set_xlabel('Total_2019', fontsize=14)
# ax.set_ylabel('Frequency', fontsize=14)

# 自定义坐标轴刻度字体大小
ax.tick_params(axis='both', labelsize=12)

# 设置y轴刻度范围为0-3000
# ax.set_ylim(0, 3000)

# 显示图形
plt.show()
```
